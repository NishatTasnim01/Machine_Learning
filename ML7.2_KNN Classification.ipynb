```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('cancer.xlsx')
```

# Label Encoder

```python
df.columns
```

```python
from pandas.api.types import is_numeric_dtype
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.columns:
    if not is_numeric_dtype(df[col]):
        df[col] = le.fit_transform(df[col])

df.head()
```

# Separate X,Y

```python
df.drop('id', axis=1,inplace=True)
```

```python
df.head()
```

```python
x = df.drop('diagnosis',axis=1)
x.head()
```

```python
y = df[['diagnosis']]
y.head()
```

# Train & Test

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest = tts(x,y,test_size=0.3,random_state=2)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

# Applying Decision Tree Classifier 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(xtrain, ytrain)
```

```python
plt.figure(figsize=(20, 20))
tree.plot_tree(dt, filled=True, rounded=True, feature_names=x.columns)
plt.savefig('d_tree.png', dpi=1200)
```

```python
dt_pred = dt.predict(xtest)
```

```python
dt_pred
```

# Accuracy, MSE, MAE for Decision Tree Classifier

```python
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
```

```python
dt_accuracy = accuracy_score(ytest, dt_pred)
print("Accuracy:", dt_accuracy)
```

```python
dt_mse = mean_squared_error(ytest, dt_pred)
print("MSE:", dt_mse)
```

```python
dt_mae = mean_absolute_error(ytest, dt_pred)
print("MAE:", dt_mae)
```

# Applying KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
ne = KNeighborsClassifier()
```

```python
ne.fit(xtrain,ytrain)
```

```python
knn_pred = ne.predict(xtest)
knn_pred
```

# Evaluate Model For Decision Tree

```python
from sklearn.metrics import confusion_matrix, roc_curve, classification_report, roc_auc_score
```

```python
print(classification_report(ytest,dt_pred))
```

```python
con_matrix = confusion_matrix(ytest,dt_pred)
con_matrix
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(con_matrix,annot=True)
plt.xlabel('Predicted Result')
plt.ylabel('Actual Value')
```

```python
y_pred_proba_dt = dt.predict_proba(xtest)[:, 1]
fpr_dt, tpr_dt, thresholds_dt = roc_curve(ytest, y_pred_proba_dt)
auc_dt = roc_auc_score(ytest, y_pred_proba_dt)

plt.plot(fpr_dt, tpr_dt, label='Decision Tree (AUC = {:.2f})'.format(auc_dt))
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Decision Tree')
plt.legend()
plt.show()
```

# Parameter Tuning for Decision Tree

Parameter Tuning Using Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid1 = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}
```

```python
grid_search1 = GridSearchCV(dt, param_grid1, cv=5, scoring='accuracy')
```

```python
grid_search1.fit(xtrain, ytrain)
```

```python
best_params1 = grid_search1.best_params_
print("Best Parameters:", best_params1)
```

```python
best_score1 = grid_search1.best_score_
print("Best Score:", best_score1)
```

Parameter Tuning Using Randomized Search

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
param_dist1 = {
    'criterion': ['gini', 'entropy'],
    'max_depth': randint(1, 20),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 10),
    'max_features': ['sqrt', 'log2', None]
}
```

```python
random_search1 = RandomizedSearchCV(estimator=dt, param_distributions=param_dist1, n_iter=10, cv=5, scoring='accuracy')
```

```python
random_search1.fit(xtrain, ytrain)
```

```python
best_params2 = random_search1.best_params_
print("Best Parameters:", best_params2)
```

```python
best_score2 = random_search1.best_score_
print("Best Score:", best_score2)
```

Parameter Tuning Using Randomized Search
